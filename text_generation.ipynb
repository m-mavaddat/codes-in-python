 بارگذاری مدل و توکنایزر

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

آماده‌سازی ورودی (پرومپت)

```python
prompt = "In the future, AI will"
input_ids = tokenizer.encode(prompt, return_tensors="pt")
```

 تولید متن

```python
output = model.generate(
    input_ids,
    max_length=100,
    num_return_sequences=1,
    do_sample=True, # برای ایجاد تنوع
    top_k=50, # فقط ۵۰ کلمه محتمل را در نظر بگیر
    top_p=0.95, # nucleus sampling
    temperature=0.9, # کنترل خلاقیت (پایین‌تر = محافظه‌کارتر)
)

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

استفاده از مدل‌های فارسی

```python
model_name = "HooshvareLab/gpt2-fa"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

prompt = "در آینده، هوش مصنوعی"
input_ids = tokenizer.encode(prompt, return_tensors="pt")

output = model.generate(input_ids, max_new_tokens=50, do_sample=True)
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```
